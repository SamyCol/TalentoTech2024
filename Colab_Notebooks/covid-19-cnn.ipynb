```python
import tensorflow as tf
import matplotlib.pyplot as plt
from keras.layers import *
from keras.models import Sequential
from keras.preprocessing import image
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import os
import warnings
warnings.filterwarnings('ignore')
```

# Read Dataset

```python
train_path = '/kaggle/input/covid19-image-dataset/Covid19-dataset/train'
test_path = '/kaggle/input/covid19-image-dataset/Covid19-dataset/test'
```

## Data Classes


```python
data_classes = os.listdir('/kaggle/input/covid19-image-dataset/Covid19-dataset/train')
print('Classes', data_classes)
```

# Preprocessing

```python
data_generator = ImageDataGenerator(rescale = 1/255)
```

```python
#train datagenerator

train_data = data_generator.flow_from_directory(train_path,
                                                target_size = (200, 200),
                                                batch_size = 16
                                               )
```

```python
# test datagenerator
test_data = data_generator.flow_from_directory(test_path,
                                              target_size=(200,200),
                                              batch_size =1)
```

# Model Building : CNN

```python
model = Sequential()

# Input Layer
model.add(Conv2D(filters = 32, kernel_size = 3, padding = 'same',
                activation ='relu' , input_shape = [200,200,3]))

# Pooling Layer --> Dimensionality Reduction
model.add(MaxPooling2D(pool_size = (2,2)))

model.add(Conv2D(filters =64, kernel_size =3, padding = 'same',
                 activation ='relu'))

model.add(MaxPooling2D(pool_size = (2,2)))

model.add(Conv2D(filters = 128, kernel_size =3, padding = 'same',
                 activation ='relu'))

model.add(MaxPooling2D())

# Flatten Layer --> 3D to 1D
model.add(Flatten())

# Adding Fully Connected Layers
model.add(Dense(128, activation ='relu'))
model.add(Dense(128, activation ='relu'))

# Output Layer
model.add(Dense(3, activation = 'softmax'))

```

```python
model.summary()
```

# Model Compilation

```python
model.compile(optimizer = 'adam',
             loss = 'categorical_crossentropy',
             metrics = ['accuracy'])
```

```python
#Fit Model
model_history = model.fit(train_data,
                         epochs = 4,
                         validation_data = test_data)
```

# Visualizing Model Results

```python
plt.plot(model_history.history['loss'], label ='Train Loss')
plt.plot(model_history.history['val_loss'], label = 'Validation Loss')
plt.legend()
plt.show()
```

```python
plt.plot(model_history.history['accuracy'], label = 'Train Accuracy')
plt.plot(model_history.history['val_accuracy'], label = 'Validation Accuracy')
plt.legend()
plt.show()
```

# Model Evaluation:

```python
model.evaluate(test_data)
```

# Save Model:

```python
model.save('covid19_model.h5')
```

# Summary

* The model is structured to progressively extract features from the input images using convolutional and pooling layers, followed by fully connected layers to make predictions based on those features.

* Purpose: This architecture is typical for image classification tasks, particularly in medical image analysis where different diseases are classified based on image patterns.

* The combination of convolutional layers helps to learn spatial hierarchies of features, making it effective for tasks like detecting pneumonia or Covid from X-ray images.
