```python
import numpy as np
import pandas as pd
import sklearn.metrics
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, ConfusionMatrixDisplay # Import ConfusionMatrixDisplay
from sklearn.metrics import confusion_matrix

matrizc = np.column_stack([[162,11],[4,22]])
print(matrizc)

"TN=162, FP=4, FN=11, TP=22"
TN = matrizc[0][0]
FP = matrizc[0][1]
FN = matrizc[1][0]
TP = matrizc[1][1]
#print("TN = ", TN)
#print("FP = ", FP)
#print("FN = ", FN)
#print("TP = ", TP)

# visulaizar matriz de confusion
vis = sklearn.metrics.ConfusionMatrixDisplay(matrizc)
vis.plot()
```

Accuarasy

```python
# accurasy = (TP+TN/SUMA)
accurasy = (TP+TN)/(TP+TN+FP+FN)
print("Acurracy = ", round(accurasy, 2))
```

Precision

```python
# precision = (TP/(TP+FP))
precision2 = (TN/(TN+FN))
print("Precision2 = ", round(precision2, 2))
precision1 = (TP/(TP+FP))
print("Precision1 = ", round(precision1, 2))

# sklearn.metrics.precision_score(y_true, y_pred, average='binary')
```

```python
# recall = TP = (TP/(FN+TP))
recall2 = (TN/(FP+TN))
print("Recall2 = ", round(recall2, 2))
recall1 = (TP/(FN+TP))
print("Recall1 = ", round(recall1, 2))
# sklearn.metrics.recall_score(y_true, y_pred, average='binary')
```
