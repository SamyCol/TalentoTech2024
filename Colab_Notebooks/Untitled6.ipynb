```python
from os import ST_RDONLY
import pandas as pd
import matplotlib.pyplot as plt

# Cargue de los datos
df = pd.read_csv('Ejercicio _Reg_Lineal.csv')  # Reemplazar 'ruta_del_archivo.csv' con la ubicación del archivo

# Paso 1: Agregar columna "xy" que multiplica los valores de "X" por "Y"
df['xy'] = df['X'] * df['Y']

# Paso 2: Agregar columna "xx" que multiplica "X" por "X"
df['xx'] = df['X'] * df['X']

# Paso 3: Calcular la suma de los valores en las columnas "xy" y "xx"
sum_xy = df['xy'].sum()
sum_xx = df['xx'].sum()

# Paso 4: Calcular el número de datos en la columna "X"
n = len(df['X'])

# Paso 5: Calcular la suma de los valores en las columnas "X" y "Y"
sum_x = df['X'].sum()
sum_y = df['Y'].sum()

# Paso 6: Calcular la variable "m" con la fórmula dada
m = (n * sum_xy - sum_x * sum_y) / (n * sum_xx - sum_x**2)

# Paso 7: Calcular la media de la columna "X" y "Y"
mean_x = df['X'].mean()
mean_y = df['Y'].mean()

# Paso 8: Calcular interceptor "b" con la fórmula dada b=mean_y - m * mean_x
df['b'] = mean_y - m * mean_x

# Paso 9: Calcular la variable "New_Y" con la fórmula dada
df['New_Y'] = m * df['X'] + df['b']

# Paso 11: Calcular Sr = suma de (y-New_Y)^2 con la fórmula dada
Sr=((df["Y"]-df["New_Y"])**2).sum()
# Mostrar resultados
print("DataFrame actualizado:\n", df)
print("\nSumatoria de la columna 'x':", sum_x)
print("\nSumatoria de la columna 'y':", sum_y)
print("\nSumatoria de la columna 'xy':", sum_xy)
print("Sumatoria de la columna 'xx':", sum_xx)
print("\nVariable 'm':", m)
print("\nVariable 'b':", df['b'][0])
print("Media de 'X':", mean_x)
print("Media de 'Y':", mean_y)
print("Sr =", mean_y)
print('y= ', m, 'x + ', df['b'][0])

plt.figure()
plt.scatter(df['X'], df['Y'], marker="X", color ="blue")
plt.plot(df['X'], df['New_Y'], color='red')
plt.show()
```
