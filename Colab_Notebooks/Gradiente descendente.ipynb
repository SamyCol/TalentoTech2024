# Gradiente descendente
# (X)^2+2x-12

```python
import numpy as np
import matplotlib.pyplot as plt

# funcion a optimzar
funcion = lambda x: (x**2)+2*x-12

# derivada de la funcion
derivada = lambda x: 2*x+2

# Rango de valores a graficar
x_values = np.linspace(-50, 50, 50)

# Grafica la funci√≥n}
plt.plot(x_values, funcion(x_values), label='f(x)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.show()

```

```python
# Definir la funcion de desecenso del gradiente
def gradient_descent(x_inicial, precision, learning_rate, funcion, derivada): # Pass funcion and derivada as arguments
    x_list, y_list = [x_inicial], [funcion(x_inicial)] # Initialize x_list and y_list separately
    while True:
        x_nuevo = x_list[-1] - learning_rate * derivada(x_list[-1]) # Use x_list[-1] for calculations
        if abs(x_nuevo - x_list[-1]) < precision:
            break
        x_list.append(x_nuevo) # Append to x_list
        y_list.append(funcion(x_nuevo)) # Append to y_list
    return x_list, y_list # Return x_list and y_list

# Define precision and learning rate before calling the function
precision = 0.0001  # Example value for precision
learning_rate = 0.1 # Example value for learning rate


# Call the gradient_descent function to get the results
# Pass funcion and derivada to the function call
x, y_list = gradient_descent(x_inicial=-30, precision=precision, learning_rate=learning_rate, funcion=funcion, derivada=derivada)

print("Precision = " + str (precision))
print("Learning rate = ", learning_rate)
print("X = ", x)
print("Y = ", y_list)
print("Minimo local en Y = ", min(y_list))
print("Minimo global en X = ", min(x))
print("numero de pasos =" + str(len(x)))
plt.scatter(x, y_list, color='red', label='Descenso del gradiente')
plt.plot(x_values, funcion(x_values), label='f(x)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.show()
```
