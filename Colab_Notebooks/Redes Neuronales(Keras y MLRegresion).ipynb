```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np # Import the numpy library and assign it the alias 'np'
from sklearn.metrics import r2_score, accuracy_score
import warnings
warnings.filterwarnings('ignore')

# Load the dataset.
df = pd.read_csv('multiple_linear_regression_dataset.csv')
df
```

```python
# Separar las variables  X1 y X2 (variables independientes) y Y(variable dependientes)
X= df[['age', 'experience']]
Y= df['income']

# Dividir datos de entrebnamiento y prueba
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# Normalizar las caracteristicas (opcion recimendable)
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
import tensorflow as tf
from tensorflow import keras

#crear modelo de regresion lineal
modelo = keras.Sequential()
modelo.add(keras.layers.Dense(units=1, input_shape=[2]))
modelo.add(keras.layers.Dense(units=3))
modelo.add(keras.layers.Dense(units=1))
modelo.summary()

# Compliar el modelo
modelo.compile(optimizer='adam', loss='mean_squared_error')

# entrenar modelo
history = modelo.fit(X_train, Y_train, epochs=100, verbose=0)

#Evaluar modelo
loss = modelo.evaluate(X_test, Y_test)
print(f'Loss o MSE en el conjunto de prueba: {loss}')
```
