```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import os
import seaborn as sns

sns.set(font_scale=1.5, style="whitegrid")
```

```python
df = np.genfromtxt('data_file.txt',delimiter="\t")
```

```python
mean = np.mean(df[:,6])
std=np.std(df[:,6])
```

```python
from scipy.stats import norm
plt.plot(df[:,6], norm.pdf(df[:,6], mean+std), label='μ + 1σ')
plt.plot(df[:,6], norm.pdf(df[:,6], mean+2*std), label='μ + 2σ')
plt.legend()
```

```python
df = pd.read_csv('data_file.txt', sep="\t")
df.columns
```

```python
def dateTime(x):
    x = x.split()
    return x[0][:4]+"-"+x[0][4:6]+"-"+x[0][6:8]+" " +x[1]
df["date and time"]=df["DATE"].apply(dateTime)
```

```python
df.info()
```

```python
fig, ax = plt.subplots(figsize=(10, 10))

# Add x-axis and y-axis
ax.scatter(df.index.values,
        df["HPCP(cm)"],
        color='purple')

# Set title and labels for axes
ax.set(xlabel="Date",
       ylabel="Precipitation (inches)",
       title="Daily Total Precipitation")

plt.show()
```
