#**ESTE ES UN EJERCICIO QUE DEPENDIENDO DE LA HORA DEL DIA QUE INGRESES TE DEVUELVE UN SALUDO**

```python
# Obtener la hora actual
print("La hora y los minutos se ingresan de manaera separada")
hora_actual = int(input("Ingrese la hora actual en formato 24 horas: "))

if hora_actual < 0 or hora_actual > 23: # Changed hora to hora_actual
    print("La hora ingresada no es válida. Por favor, ingrese una hora entre 0 y 23.")
    #hora_actual = int(input("Ingrese la hora actual en formato 24 horas: "))
else:
    minutos = int(input("Ingrese los minutos: "))
    if minutos < 0 or minutos > 59:
        print("Los minutos ingresados no son válidos. Por favor, ingrese minutos entre 0 y 59.")
        #minutos = int(input("Ingrese los minutos: "))
    else:
        minutos = minutos / 60
        hora_actual = hora_actual + minutos # Changed hora to hora_actual
# Función para generar el saludo según la hora del día
def saludo_según_hora(hora):
    if 1 <= hora < 12:
        return "Hola, buenos días"
    elif 12 <= hora < 18:
        return "Hola, buenas tardes"
    else:
        return "Hola, buenas noches"

# Generar saludo según la hora actual
saludo = saludo_según_hora(hora_actual)
saludo
```
