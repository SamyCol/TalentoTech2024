```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt

# Load the Iris dataset
iris = load_iris()
X = iris.data
X

# Perform DBSCAN clustering
dbscan = DBSCAN(eps=0.5, min_samples=5)
clusters = dbscan.fit_predict(X)

# Adiciona la etiqueta cluster al dataset
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)
iris_df['cluster'] = clusters
iris_df
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler


# Seleccionar las columnas que quieres escalar (features)
features = ['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)']
X = iris_df[features]

# Crea un objeto StandardScaler
scaler = StandardScaler()

# Ajusta el scaler a tus datos y transforma las features
X_scaled = scaler.fit_transform(X)

# Crea un nuevo DataFrame con las features escaladas y la columna 'cluster'
scaled_iris_df = pd.DataFrame(X_scaled, columns=features, index=iris_df.index)
scaled_iris_df['cluster'] = iris_df['cluster']

# Ahora 'scaled_iris_df' contiene tus datos escalados
# Puedes usar este DataFrame para visualizar o entrenar modelos.

# Ejemplo de visualizaciÃ³n con los datos escalados:
from matplotlib import pyplot as plt

scaled_iris_df.plot(kind='scatter', x='sepal length (cm)', y='petal length (cm)', c='cluster', colormap='viridis')
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.title('DBSCAN Clustering of Iris Dataset (Scaled Data)') # AÃ±ade un tÃ­tulo
plt.show()
```

```python
# Visualize the clusters
plt.scatter(iris_df['sepal length (cm)'], iris_df['sepal width (cm)'], c=iris_df['cluster'])
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('DBSCAN Clustering of Iris Dataset')
plt.show()
```
